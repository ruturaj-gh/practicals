```python
sent='Hello friends! How are you? We will be learning Python today.'
```

```python
def clean_text(sent):
    tokens=word_tokenize(sent)
    #if the whole word is either digit or alphabet then it is valid
    clean=[word for word in tokens
      if word.isdigit() or word.isalpha()]
    clean=[ps.stem(word) for word in clean
      if word not in swords]
    
    return clean
```

```python
clean
```

```python
clean_text(sent)
```

```python
#tf idf vectorizer is the method for identifying the number of times that word or phrase appears in file
```

```python
#pre-processing 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(analyzer=clean_text)
```

```python
x=df['text']
y=df['label']
```

```python
x_new=tfidf.fit_transform(x)
```

```python
#cross radiation
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
x_new,y,random_state=0,test_size=0.25)
```

```python
x_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
```

```python
nb.fit(x_train.toarray(),y_train)
```

```python
y_pred=nb.predict(x_test.toarray())
```

```python
y_test.value_counts()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=20)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
from sklearn.linear_model import LogisticRegression 
log=LogisticRegression()
log.fit(x_train,y_train)
y_pred=log.predict(x_test)
accuracy_score(y_test,y_pred)
```
